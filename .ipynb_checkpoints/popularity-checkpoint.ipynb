# Calculate a weighted rating

```python
# WR = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C 
# v - number of votes for a movie
# m - minimum number of votes required
# R - average rating of the movie
# C - average rating across all movies
```

```python
m = movies['vote_count'].quantile(0.9)
m
```
